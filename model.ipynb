```
# ストップワードの除去：使用頻度の高い言葉を処理対象外にする
import MeCab

def token_generator(text):
    # mecab-ipadic-NEologd辞書指定してオブジェクト生成
    tagger = MeCab.Tagger()
    tagger.parse("")
    # 形態素解析の結果をリストで取得、単語ごとにリストの要素に入ってる
    node = tagger.parseToNode(text)
    result = []
    
    num = 0
    oldhinshi = "" # 一つ前のトークンの品詞

    #助詞や助動詞は拾わない
    while node is not None:
        # 品詞情報取得
        # Node.featureのフォーマット：品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用形,活用型,原形,読み,発音
        hinshi = node.feature.split(",")[0]
        if hinshi in["動詞","形容詞"]:
            # 形態素情報から原形情報取得
            result.append(node.feature.split(",")[6])
            oldhinshi = hinshi # oldhinshiの更新

        elif hinshi in ["副詞"]:
            # 表層形の取得、単語の文字が入ってる
            result.append(node.surface)
            oldhinshi = hinshi # oldhinshiの更新
        
        elif hinshi in ["名詞"]:
            # 表層形の取得、単語の文字が入ってる
            if oldhinshi != '名詞': 
                result.append(node.surface)
                num += 1 # インデックス指定のためにnumで番号を振っておく
                oldhinshi = hinshi # oldhinshiの更新
            else: # oldhinshiが名詞の場合
                result[num-1] = result[num-1]+node.surface # 一つ前のトークンに接続
                oldhinshi = hinshi # oldhinshiの更新
        else:
            oldhinshi = hinshi # oldhinshiの更新
            
        node = node.next

    return result
 
```

```
import numpy as np

label = df_["s"]
label = [float(i) for i in label]
label = np.array(label)




# 特殊なトークンは事前に定義しておく
PAD_TOKEN = '<PAD>'  # バッチ処理の際に、短い系列の末尾を埋めるために使う （Padding）
UNK_TOKEN = '<Unknown>'  # 語彙に存在しない単語を表す （Unknown）
PAD = 0
UNK = 1


# In[20]:


MIN_COUNT = 2  # 語彙に含める単語の最低出現回数

# 単語をIDに変換する辞書の初期値を設定
word2id = {
    PAD_TOKEN: PAD,
    UNK_TOKEN: UNK,
    }


# In[21]:


class Vocab(object):
    def __init__(self, word2id={}):
        """
        word2id: 単語(str)をインデックス(int)に変換する辞書
        id2word: インデックス(int)を単語(str)に変換する辞書
        """
        self.word2id = dict(word2id)
        self.id2word = {v: k for k, v in self.word2id.items()}    
        
    def build_vocab(self, sentences, min_count=1):
        # 各単語の出現回数の辞書を作成する
        word_counter = {}
        for sentence in sentences:
            for word in sentence:
                word_counter[word] = word_counter.get(word, 0) + 1

        # min_count回以上出現する単語のみ語彙に加える
        for word, count in sorted(word_counter.items(), key=lambda x: -x[1]):
            if count < min_count:
                break
            _id = len(self.word2id)
            self.word2id.setdefault(word, _id)
            self.id2word[_id] = word 


vocab_Y = Vocab(word2id=word2id)
vocab_Y.build_vocab(fulltoken, min_count=MIN_COUNT)
vocab_size_Y = len(vocab_Y.id2word)

```

```
def sentence_to_ids(vocab, sentence):
    # 単語(str)のリストをID(int)のリストに変換する関数
    ids = [vocab.word2id.get(word, UNK) for word in sentence]
    return ids


# In[25]:


data_x_vec = [sentence_to_ids(vocab_Y, sentence) for sentence in fulltoken]
#data_x_vec[:2]


# In[26]:


# 文章の長さを揃えるため、0でパディングする
max_sentence_size = 0
for sentence_vec in data_x_vec:
    if max_sentence_size < len(sentence_vec):
        max_sentence_size = len(sentence_vec)
#print(max_sentence_size)   # ==len(sentence_ids)

for sentence_ids in data_x_vec:
    while len(sentence_ids) < max_sentence_size:
         # 末尾に0を追加
        sentence_ids.append(0)


# In[27]:


# arrayに変換
data_x_vecs = np.array(data_x_vec, dtype="int32")
#data_x_vecs



# 訓練データとテストデータにわけます
from sklearn.model_selection import train_test_split
train_x_, test_x, train_y_, test_y = train_test_split(data_x_vecs, label, test_size = 0.1)



# 訓練データとテストデータにわけます
from sklearn.model_selection import train_test_split
train_x, valid_x, train_y, valid_y = train_test_split(train_x_, train_y_, test_size = 0.2)


# In[33]:


```

```
# オプティマイザー：Adam、損失関数：（多クラス）cross entropy
model.compile(optimizer='Adam', loss='mean_squared_error')

# train
model.fit(train_x, train_y, epochs=30, batch_size=256)
```
